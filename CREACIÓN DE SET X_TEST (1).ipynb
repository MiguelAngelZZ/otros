```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
kernel='C:/Users/miavila/Documents/U/PROYECTO P/variables exogenas/'
festivos=pd.read_excel(kernel +'lista_festivos 2022.xlsx' )
festividades=pd.read_excel(kernel +'lista_festividades_2022.xlsx' )
```

# Set por hora - para cant. llamadas

```python
def etl():
    #GENERACIÓN DE FECHA 
    gen=np.arange('2022-09-01', '2023-09-30', dtype='datetime64[h]')
    dfgen=pd.DataFrame(gen)
    df=dfgen.copy()
    df.rename(columns={0: "FechaT"},inplace=True)

    #CREAR COLUMNAS CORRESPONDIENTES A MES, DIA DEL MES, DIA DE LA SEMANA (INGLES)
    dfa=df.assign(MesN=df.FechaT.dt.month   ,Dia_MesN=df.FechaT.dt.day    ,Dia_SemN=df.FechaT.dt.day_name()) 

    # RENOMBRAR DIAS DE LA SEMANA POR NUMEROS
    #['Sábado', 'Domingo', 'Lunes', 'Martes', 'Miércoles', 'Jueves','Viernes'],[6,7,1,2,3,4,5]
    dfa['Dia_SemN'].replace(['Thursday', 'Friday', 'Saturday', 'Sunday', 'Monday', 'Tuesday','Wednesday']
    ,[4,5,6,7,1,2,3],inplace=True)  
    
    #extrae solo año dia mes   columna solo fecha
    dfa['Fecha'] = dfa['FechaT'].dt.date.astype(str)
    dfa['Fecha'] =pd.to_datetime(dfa['Fecha'])
    #CREAR COLUMNA CON LA HORA
    dfa['Hora24']=dfa['FechaT'].dt.hour.astype(str) #df['Hora24']=pd.DatetimeIndex(df.FechaT).hour
    

    # INDEXAR FESTIVOS
    df1 = pd.merge(dfa,festivos,on='Fecha',how='left')
    # INDEXAR FESTIVIDADES
    dfa = pd.merge(df1,festividades,on='Fecha',how='left')

    # ordenar
    dfa=pd.DataFrame(dfa,columns=['FechaT','Fecha',
    'festivo', 'festividades', 'Dia_SemN','Dia_MesN', 'MesN','Hora24'])

    # guarda en excel
    dfa.to_excel(kernel+ 'X_new_hour.xlsx')
    
    return(dfa)
pass

DF=pd.DataFrame(etl());DF
```
