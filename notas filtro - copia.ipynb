```python
import pandas as pd
import numpy as np
```

```python
from pandas import read_excel


notas=pd.read_excel("C:/Users/miavila/Desktop/notas coordis/Notas consolidadas 2022.xlsx",sheet_name='CUARTILES')
```

```python
notas
```

```python
NT=pd.DataFrame(notas,columns=['COORD','Mes','CUARTIL EFECTIVIDAD','CUARTIL NPS', 'CUARTIL SERVICIO'])
```

```python
NT
```

```python
agru1 = NT.groupby([NT.COORD,NT.Mes,NT['CUARTIL EFECTIVIDAD']])['COORD'].count()
agru2= NT.groupby([NT.COORD,NT.Mes,NT['CUARTIL NPS']])['COORD'].count()
agru3=  NT.groupby([NT.COORD,NT.Mes,NT['CUARTIL SERVICIO']])['COORD'].count()
```

```python
agru1
```

```python
agru11 = agru1.index.to_frame(name=['COORD','Mes','CUARTIL EFECTIVIDAD'])
agru22 = agru2.index.to_frame(name=['COORD','Mes','CUARTIL NPS'])
agru33 = agru3.index.to_frame(name=['COORD','Mes','CUARTIL SERVICIO'])
```

```python
agru33
```

```python
agru11.index = range(agru11.shape[0])
agru22.index = range(agru22.shape[0])
agru33.index = range(agru33.shape[0])
```

```python
DF1=pd.DataFrame(agru11,columns=['COORD','Mes','CUARTIL EFECTIVIDAD'])
DF2=pd.DataFrame(agru22,columns=['COORD','Mes','CUARTIL NPS'])
DF3=pd.DataFrame(agru33,columns=['COORD','Mes','CUARTIL SERVICIO'])
```

```python
DF1,DF2,DF3
```

```python
recib1=pd.DataFrame(agru1.values,columns=['Count'])
recib2=pd.DataFrame(agru2.values,columns=['Count'])
recib3=pd.DataFrame(agru3.values,columns=['Count'])
```

```python
recib1
```

```python
DF11=pd.concat([DF1,recib1], axis=1,)
DF22=pd.concat([DF2,recib2], axis=1,)
DF33=pd.concat([DF3,recib3], axis=1,)
```

```python
DF11
```

```python

with pd.ExcelWriter('CONT_NOTAS COORDINADORES 2022.xlsx') as writer:  
    DF11.to_excel(writer, sheet_name='Q EFECTIVIDAD')
    DF22.to_excel(writer, sheet_name='Q NPS')
    DF33.to_excel(writer, sheet_name='Q SERVICIO')
```

```python

with pd.ExcelWriter('CONT NOTAS T.xlsx') as writer:  
    DF11.T.to_excel(writer, sheet_name='Q EFECTIVIDAD')
    DF22.T.to_excel(writer, sheet_name='Q NPS')
    DF33.T.to_excel(writer, sheet_name='Q SERVICIO')
```
