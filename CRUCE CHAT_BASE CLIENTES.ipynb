```python
clien=pd.read_excel('C:/Users/miavila/Desktop/CRUCE SCORE/DICIEMBRE/base_clientes_banco.csv.xlsx')
clien=pd.DataFrame(clien,columns=['nro_identificacion', 'cliente_bancario', 'segmento_digital',
       'score_integral'])
```

```python
clien
```

```python
clien.dropna(inplace=True)
```

```python
#
clien.rename(columns={'nro_identificacion':'Cedula'},inplace=True)
clien['Cedula']=clien['Cedula'].astype(int)
```

```python
clien.drop_duplicates(subset=['Cedula'],keep='last', inplace=True)
```

```python
CHAT.Cedula.replace(np.nan,-1,inplace=True)
```

```python
import re 
def extraer_tratamiento(name):
  m = re.search(r"\w+\.",str(name))
  if not m:
      return -1 # CON CARATERES *
  return m[0][:-1]
```

```python
CHAT['Cedula']=CHAT['Cedula'].apply(extraer_tratamiento)
```

```python
CHAT['Cedula']=CHAT['Cedula'].astype(np.int64)
```

```python


nueva_fila2 = {'Cedula':-1, 'cliente_bancario':np.nan, 'segmento_digital':np.nan, 'score_integral':np.nan}

#Añadiendo una fila al dataframe
clien =clien.append(nueva_fila2, ignore_index=True)
```

```python
df=pd.merge(CHAT,clien,on=['Cedula'])
```
