## Filtro para palabra especifica (mas flexible) en motivos o producto


+  str.contains()     una cadena de texto específica
+  str.startswith()   una cadena de texto específica al principio de una columna 
+  str.endswith()     una cadena de texto específica al final de una columna
+  str.match()        busca un patrón de texto específico dentro de una columna


Todas siguen los mismos PARAMETROS:


+  pat (str): La cadena de texto que se busca en la columna.
+ case (bool): Si es True, la búsqueda será sensible a mayúsculas y minúsculas.
 Si es False, la búsqueda no tendrá en cuenta mayúsculas y minúsculas.
+ na (bool): Si es True, los valores NaN se tratarán como coincidencias. Si es False, los valores NaN no se tendrán en cuenta.


________________________________________________________________________
+ flags (int): Una bandera para indicar cómo se realiza la búsqueda.
  
  + re.IGNORECASE: Esta bandera indica que la búsqueda no tendrá en cuenta mayúsculas y minúsculas.
  + re.MULTILINE: Esta bandera indica que la búsqueda se realizará en múltiples líneas.
  + re.DOTALL: Esta bandera indica que el punto (.) coincidirá con cualquier carácter.
  + re.UNICODE: Esta bandera indica que la búsqueda se realizará con caracteres Unicode.
  + re.VERBOSE: Esta bandera indica que la búsqueda se realizará con una sintaxis más legible.


```python
fil=agru1[agru1
  ['Motivos']
  .str.contains('proDuCTo', case=False, na=True)]
fil
```
