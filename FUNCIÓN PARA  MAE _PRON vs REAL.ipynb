```python

#crea una función que itere entre dias y arroje al final una lista con el valor real, el pronostico, el MAE Y RMSE, TENIENDO EN CUENTA QUE SON 24 DATOS POR DIA

def calcular_MAE_RMSE(df):
    #inicializar listas para almacenar los resultados
    valor_real = []
    pronostico = []
    mae = []
    rmse = []
    
    #iterar sobre los días
    for dia in df['dia'].unique():
        #obtener los datos para el día actual
        df_dia = df[df['dia'] == dia]
        
        #calcular el MAE para el día actual
        mae_dia = np.mean(np.abs(df_dia['pronostico'] - df_dia['valor_real']))/len(df_dia)
        
        #calcular el RMSE para el día actual
        rmse_dia = np.sqrt(np.mean((df_dia['pronostico'] - df_dia['valor_real'])**2))/len(df_dia)
        
        #agregar los resultados a las listas
        valor_real.extend(list(df_dia['valor_real']))
        pronostico.extend(list(df_dia['pronostico']))
        mae.append(mae_dia)
        rmse.append(rmse_dia)
        
    #devolver los resultados en un dataframe
    df_resultados = pd.DataFrame({'valor_real': valor_real, 'pronostico': pronostico, 'mae': mae, 'rmse': rmse})
    return df_resultados
```

```python
#crear función que itere entre años y por separado tambien para años y arroje al final una lista con el valor real, el pronostico, el MAE Y RMSE, TENIENDO EN CUENTA QUE SON 12 DATOS POR AÑO

def calcular_MAE_RMSE_anio(df):
    #inicializar listas
    lista_valor_real = []
    lista_pronostico = []
    lista_mae = []
    lista_rmse = []
    
    #agrupar los datos por año
    df_anio = df.groupby('anio')
    
    #iterar sobre los grupos
    for nombre, grupo in df_anio:
        #calcular MAE
        mae = np.mean(np.abs(grupo['pronostico'] - grupo['valor_real']))/len(grupo)
        #calcular RMSE
        rmse = np.sqrt(np.mean((grupo['pronostico'] - grupo['valor_real'])**2))/len(grupo)
        #agregar los datos a las listas
        lista_valor_real.extend(grupo['valor_real'].tolist())
        lista_pronostico.extend(grupo['pronostico'].tolist())
        lista_mae.extend([mae]*len(grupo))
        lista_rmse.extend([rmse]*len(grupo))
    
    #crear dataframe con los datos
    df_resultados = pd.DataFrame({'valor_real':lista_valor_real, 'pronostico':lista_pronostico, 'mae':lista_mae, 'rmse':lista_rmse})
    
    return df_resultados
```
